```python
import numpy as np
import matplotlib.pyplot as plt
from pspy import so_spectra

spec_dir = '/scratch/gpfs/SIMONSOBS/users/zatkins/projects/lat-iso/piso/tiger_dr6xdeep56_20251119/spectra/'
maps = ['i1_f090', 'i1_f150', 'i3_f090', 'i3_f150', 'i4_f090', 'i4_f150', 'c1_f220', 'c1_f280', 'i5_f220', 'i5_f280']
spec_file_name = 'Dl_so_{ma}xso_{mb}_noise.dat'

spectra = ['TT', 'TE', 'TB', 'ET', 'BT', 'EE', 'EB', 'BE', 'BB']
```

```python
# get correlations between bands on the same tube for T E B autos
tubes2freqs = {}
for m in maps:
    tube, freq = m.split('_')
    if tube not in tubes2freqs:
        tubes2freqs[tube] = [freq]
    else:
        tubes2freqs[tube].append(freq)

for tube, freqs in tubes2freqs.items():
    _, ps_dict0 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=f'{tube}_{freqs[0]}', mb=f'{tube}_{freqs[0]}'), spectra=spectra)
    _, ps_dict1 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=f'{tube}_{freqs[1]}', mb=f'{tube}_{freqs[1]}'), spectra=spectra)
    lb, ps_dictx = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=f'{tube}_{freqs[0]}', mb=f'{tube}_{freqs[1]}'), spectra=spectra)

    for spec in ['TT', 'EE', 'BB']:
        corr = ps_dictx[spec] / np.sqrt(ps_dict0[spec] * ps_dict1[spec])
        plt.plot(lb, corr, label=f'{tube} {freqs[0]}x{freqs[1]} {spec} corr')
    plt.legend()
    plt.ylim(-.25, 1)
    plt.grid()
    plt.show()
```

```python
fig, axs = plt.subplots(figsize=(36, 36), nrows=len(maps) - 1, ncols=len(maps) - 1)

# TT 
for i in range(len(maps) - 1):
    for j in range(len(maps) - 1):
        ax = axs[i, j]

        if i > j:
            ax.axis('off')
            continue

        j += 1

        _, ps_dict0 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[i]), spectra=spectra)
        _, ps_dict1 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[j], mb=maps[j]), spectra=spectra)
        lb, ps_dictx = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[j]), spectra=spectra)
        
        for spec in ['TT']:
            corr = ps_dictx[spec] / np.sqrt(ps_dict0[spec] * ps_dict1[spec])
            ax.plot(lb, corr, label=spec)

        if i == 0:
            ax.set_title(maps[j])        
        
        if j == i + 1:
            ax.set_xticks([0, 2000, 4000, 6000], [0, 2000, 4000, ''])
            ax.set_yticks([-.2, 0., .2, .4, .6, .8, 1.0], [-.2, 0., .2, .4, .6, .8, 1.0])
            ax.set_ylabel(maps[i])
        else:
            ax.set_xticks([0, 2000, 4000, 6000], [])
            ax.set_yticks([-.2, 0., .2, .4, .6, .8, 1.0], [])

        ax.legend()
        ax.set_xlim(0, 6000)
        ax.set_ylim(-.2, 1)
        ax.grid()

plt.show()

fig, axs = plt.subplots(figsize=(36, 36), nrows=len(maps) - 1, ncols=len(maps) - 1)

# TT zoom
for i in range(len(maps) - 1):
    for j in range(len(maps) - 1):
        ax = axs[i, j]

        if i > j:
            ax.axis('off')
            continue

        j += 1

        _, ps_dict0 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[i]), spectra=spectra)
        _, ps_dict1 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[j], mb=maps[j]), spectra=spectra)
        lb, ps_dictx = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[j]), spectra=spectra)
        
        for spec in ['TT']:
            corr = ps_dictx[spec] / np.sqrt(ps_dict0[spec] * ps_dict1[spec])
            ax.plot(lb, corr, label=spec)

        if i == 0:
            ax.set_title(maps[j])        
        
        if j == i + 1:
            ax.set_xticks([0, 300, 600, 900], [0, 300, 600, ''])
            ax.set_yticks([-.75, -.5, -.25, 0, .25, .5, .75], [-.75, -.5, -.25, 0, .25, .5, .75])
            ax.set_ylabel(maps[i])
        else:
            ax.set_xticks([0, 300, 600, 900], [])
            ax.set_yticks([-.75, -.5, -.25, 0, .25, .5, .75], [])

        ax.legend()
        ax.set_xlim(0, 900)
        ax.set_ylim(-.75, .75)
        ax.grid()

plt.show()

fig, axs = plt.subplots(figsize=(36, 36), nrows=len(maps) - 1, ncols=len(maps) - 1)

# EE, BB 
for i in range(len(maps) - 1):
    for j in range(len(maps) - 1):
        ax = axs[i, j]

        if i > j:
            ax.axis('off')
            continue

        j += 1

        _, ps_dict0 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[i]), spectra=spectra)
        _, ps_dict1 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[j], mb=maps[j]), spectra=spectra)
        lb, ps_dictx = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[j]), spectra=spectra)
        
        for k, spec in enumerate(['EE', 'BB']):
            corr = ps_dictx[spec] / np.sqrt(ps_dict0[spec] * ps_dict1[spec])
            ax.plot(lb, corr, label=spec, color=f'C{k+1}')

        if i == 0:
            ax.set_title(maps[j])        
        
        if j == i + 1:
            ax.set_xticks([0, 2000, 4000, 6000], [0, 2000, 4000, ''])
            ax.set_yticks([-.2, 0., .2, .4, .6, .8, 1.0], [-.2, 0., .2, .4, .6, .8, 1.0])
            ax.set_ylabel(maps[i])
        else:
            ax.set_xticks([0, 2000, 4000, 6000], [])
            ax.set_yticks([-.2, 0., .2, .4, .6, .8, 1.0], [])

        ax.legend()
        ax.set_xlim(0, 6000)
        ax.set_ylim(-.2, 1)
        ax.grid()

plt.show()

fig, axs = plt.subplots(figsize=(36, 36), nrows=len(maps) - 1, ncols=len(maps) - 1)

# EE, BB zoom
for i in range(len(maps) - 1):
    for j in range(len(maps) - 1):
        ax = axs[i, j]

        if i > j:
            ax.axis('off')
            continue

        j += 1

        _, ps_dict0 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[i]), spectra=spectra)
        _, ps_dict1 = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[j], mb=maps[j]), spectra=spectra)
        lb, ps_dictx = so_spectra.read_ps(spec_dir + spec_file_name.format(ma=maps[i], mb=maps[j]), spectra=spectra)
        
        for k, spec in enumerate(['EE', 'BB']):
            corr = ps_dictx[spec] / np.sqrt(ps_dict0[spec] * ps_dict1[spec])
            ax.plot(lb, corr, label=spec, color=f'C{k+1}')

        if i == 0:
            ax.set_title(maps[j])        
        
        if j == i + 1:
            ax.set_xticks([0, 300, 600, 900], [0, 300, 600, ''])
            ax.set_yticks([-.75, -.5, -.25, 0, .25, .5, .75], [-.75, -.5, -.25, 0, .25, .5, .75])
            ax.set_ylabel(maps[i])
        else:
            ax.set_xticks([0, 300, 600, 900], [])
            ax.set_yticks([-.75, -.5, -.25, 0, .25, .5, .75], [])

        ax.legend()
        ax.set_xlim(0, 900)
        ax.set_ylim(-.75, .75)
        ax.grid()

plt.show()
```
